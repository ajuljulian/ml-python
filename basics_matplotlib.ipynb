```python
import matplotlib.pyplot as plt
import random
```

### Line plot

```python
x = [i for i in range(20)]
y = [random.randrange(100) for i in range(20)]
plt.plot(x, y)
plt.show()
```

### Scatter plot

```python
x = [i for i in range(20)]
y = [random.randrange(100) * i for i in range(20)]
plt.scatter(x, y)
plt.show()
```

### Histograms

```python
ages = [random.randrange(100) for i in range(100)]
plt.hist(ages, bins=10)
plt.show()
```

### Bar chart

```python
import collections
letter_frequencies = collections.Counter("This is a long paragraph about anything.  Let's see if it makes sense")
print(letter_frequencies)
x = [i for i in range(len(letter_frequencies))]
plt.bar(x, letter_frequencies.values())
plt.xticks(x, letter_frequencies.keys())
plt.show()
```
